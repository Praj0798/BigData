```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
params = pd.read_csv("params.csv")
params.head()
```

```python
rank_50 = params[params["rank"] == 50]
rank_100 = params[params["rank"] == 100]
rank_150 = params[params["rank"] == 150]
rank_200 = params[params["rank"] == 200]
rank_250 = params[params["rank"] == 250]
rank_300 = params[params["rank"] == 300]
rank_350 = params[params["rank"] == 350]
rank_400 = params[params["rank"] == 400]
rank_450 = params[params["rank"] == 450]
```

```python
plt.plot(rank_50["MAP@100"], linestyle = "dotted")
```
